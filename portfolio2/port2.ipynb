## Apply Monte Carlo Simulation 

```python
import numpy as np
```

```python
num_portfolios = 10000

# Array to store the results: returns, volatility (risk), and Sharpe ratio for each portfolio
results = np.zeros((4, num_portfolios))

# Array to store the weights of each portfolio
all_weights = np.zeros((num_portfolios, len(nifty50_top_returns.columns)))

# Loop through the number of portfolios and capture weights
for i in range(num_portfolios):
    # Randomly generate portfolio weights
    weights = np.random.random(len(nifty50_top_returns.columns))
    weights /= np.sum(weights)  # Normalize weights to sum to 1

    # Store the weights
    all_weights[i, :] = weights

    # Calculate portfolio return and risk (volatility)
    portfolio_return = np.sum(weights * top_mean_returns) * 252  # 252 trading days in a year
    portfolio_stddev = np.sqrt(np.dot(weights.T, np.dot(top_cov_matrix * 252, weights)))

    # Calculate Sharpe ratio
    sharpe_ratio = (portfolio_return - risk_free_rate) / portfolio_stddev

    # Store the results
    results[0, i] = portfolio_return
    results[1, i] = portfolio_stddev
    results[2, i] = sharpe_ratio
    results[3, i] = np.sum(weights)  # Ensure weights sum to 1

# Convert the results array into a DataFrame
results_df = pd.DataFrame(results.T, columns=['Return', 'Volatility', 'Sharpe Ratio', 'Sum of Weights'])
```

### Risk Management: Value at Risk (VaR) and Conditional Value at Risk (CVaR)

### 1. Value at Risk (VaR):
Value at Risk is the maximum loss a portfolio is expected to suffer over a given time period with a certain confidence level. For example, a 95% VaR of $10,000 means that there’s a 5% chance the portfolio will lose more than $10,000 over the time horizon.

### VaR Formula:
For normally distributed returns, VaR at confidence level $\alpha$ is calculated as:

$$
VaR = Z_{\alpha} \times \sigma_p \times \sqrt{T}
$$

Where:

- $Z_{\alpha}$ is the Z-score corresponding to the confidence level.
- $\sigma_p$ is the portfolio's standard deviation (volatility).
- $T$ is the time horizon (e.g., 1 day, 10 days, etc.).

### 2. Conditional Value at Risk (CVaR):
Conditional Value at Risk (also known as Expected Shortfall) is the average loss that exceeds the VaR threshold. While VaR tells you the worst-case loss up to a certain point, CVaR provides an estimate of the average loss in the worst-case scenarios.

### CVaR Formula:
To calculate CVaR, we can take the average of the losses that exceed the VaR:

$$
CVaR = E[Loss | Loss > VaR]
$$

This is typically calculated by simulating losses and taking the average of the worst outcomes.

```python
import numpy as np

def calculate_var_cvar(returns, confidence_level=0.95):
    # Sort the returns in ascending order
    sorted_returns = np.sort(returns)
    
    # Calculate the index for VaR
    var_index = int((1 - confidence_level) * len(sorted_returns))
    
    # VaR at the specified confidence level
    var = sorted_returns[var_index]
    
    # CVaR (Expected Shortfall): average of returns below VaR
    cvar = np.mean(sorted_returns[:var_index])
    
    return var, cvar

# Run Monte Carlo simulation for portfolio returns (using already simulated portfolio returns)
portfolio_returns = results_df['Return']  # Simulated returns from previous step

# Calculate VaR and CVaR for the Maximum Sharpe Ratio portfolio (example with 95% confidence level)
confidence_level = 0.95
var, cvar = calculate_var_cvar(portfolio_returns, confidence_level)

# Display results
print(f"Value at Risk (VaR) at {confidence_level*100}% confidence level: {var}")
print(f"Conditional Value at Risk (CVaR): {cvar}")
```

- There is only a 5% chance that the portfolio will experience a loss worse than 0.7092.i.e, will not lose more than 0.7092 (risk averse)
