```python
import yfinance as yf
import pandas as pd


# Fetch historical data for each stock
def fetch_data(tickers):
    data = {}
    for ticker in tickers:
        stock_data = yf.download(ticker, start='2023-10-01', end='2024-10-01')  # Adjust date range
        data[ticker] = stock_data['Adj Close']  # Get adjusted close prices
    return pd.DataFrame(data)

# Get the data for Nifty 50 stocks
nifty50_data = fetch_data(nifty50_tickers)


# Display the data
nifty50_data.head()
```

### Calculate Mean , Variance , Covariance

```python
nifty50_returns = nifty50_data_clean.pct_change().dropna()

# Calculate mean returns, standard deviations, and covariance matrix
mean_returns = nifty50_returns.mean()
std_dev_returns = nifty50_returns.std()
cov_matrix = nifty50_returns.cov()

# Display the results
print("Mean Returns:")
print(mean_returns)

```

Can move forward with the *Monte Carlo simulation* to generate different portfolio combinations and assess their returns and risks.

### Key Components of the Simulation:

Random Portfolio Weights: Generate random weights for the stocks in the portfolio. The sum of these weights should be 100%.

Portfolio Return: Calculate the expected portfolio return using the randomly generated weights.

Portfolio Risk (Volatility): Calculate the portfolio risk (standard deviation) using the covariance matrix.

Run Simulations: Run a large number of simulations (e.g., 10,000 or more) to generate different portfolios.

Analyze Results: Capture the return, risk, and other metrics (like the Sharpe ratio) for each portfolio.

```python
import numpy as np

# Number of simulations
num_portfolios = 10000

# Array to store the results: returns, volatility (risk), and Sharpe ratio for each portfolio
results = np.zeros((4, num_portfolios))

# Risk-free rate for Sharpe ratio calculation (adjust as needed)
risk_free_rate = 0.001

# Loop through the number of portfolios
for i in range(num_portfolios):
    # Randomly generate portfolio weights
    weights = np.random.random(len(nifty50_returns.columns))
    weights /= np.sum(weights)  # Normalize weights to sum to 1

    # Calculate portfolio return and risk (volatility)
    portfolio_return = np.sum(weights * mean_returns) * 252  # 252 trading days in a year
    portfolio_stddev = np.sqrt(np.dot(weights.T, np.dot(cov_matrix * 252, weights)))

    # Calculate Sharpe ratio
    sharpe_ratio = (portfolio_return - risk_free_rate) / portfolio_stddev

    # Store the results
    results[0,i] = portfolio_return
    results[1,i] = portfolio_stddev
    results[2,i] = sharpe_ratio
    results[3,i] = np.sum(weights)  # Ensure weights sum to 1 (sanity check)

# Convert results array into a DataFrame for better readability
results_df = pd.DataFrame(results.T, columns=['Return', 'Volatility', 'Sharpe Ratio', 'Sum of Weights'])

# Display the first few results
results_df.head()
```

## Plotting the Efficient Frontier
The Efficient Frontier is a plot that shows the best possible portfolios that provide the highest return for a given level of risk. Let’s create a plot to visualize the portfolios and identify the optimal ones.

```python
import matplotlib.pyplot as plt

# Plotting the portfolios in the risk-return space
plt.figure(figsize=(10, 6))
plt.scatter(results_df['Volatility'], results_df['Return'], c=results_df['Sharpe Ratio'], cmap='viridis', marker='o')
plt.colorbar(label='Sharpe Ratio')
plt.title('Monte Carlo Portfolio Optimization')
plt.xlabel('Volatility (Risk)')
plt.ylabel('Return')

# Highlight the portfolio with the highest Sharpe ratio
max_sharpe_idx = results_df['Sharpe Ratio'].idxmax()
max_sharpe_portfolio = results_df.loc[max_sharpe_idx]

plt.scatter(max_sharpe_portfolio['Volatility'], max_sharpe_portfolio['Return'], color='r', marker='*', s=200, label='Max Sharpe Ratio')

# Highlight the portfolio with the lowest volatility (risk)
min_vol_idx = results_df['Volatility'].idxmin()
min_vol_portfolio = results_df.loc[min_vol_idx]

plt.scatter(min_vol_portfolio['Volatility'], min_vol_portfolio['Return'], color='b', marker='*', s=200, label='Min Volatility')

plt.legend(loc='upper left')
plt.show()

# Display the details of the Max Sharpe Ratio and Min Volatility portfolios
print("Portfolio with Maximum Sharpe Ratio:")
print(max_sharpe_portfolio)

print("\nPortfolio with Minimum Volatility:")
print(min_vol_portfolio)
```

## Optimal Weights

```python
# Array to store the weights of each portfolio
all_weights = np.zeros((num_portfolios, len(nifty50_returns.columns)))

# Loop through the number of portfolios and capture weights
for i in range(num_portfolios):
    # Randomly generate portfolio weights
    weights = np.random.random(len(nifty50_returns.columns))
    weights /= np.sum(weights)  # Normalize weights to sum to 1

    # Store the weights
    all_weights[i, :] = weights

    # Calculate portfolio return and risk (volatility)
    portfolio_return = np.sum(weights * mean_returns) * 252  # 252 trading days in a year
    portfolio_stddev = np.sqrt(np.dot(weights.T, np.dot(cov_matrix * 252, weights)))

    # Calculate Sharpe ratio
    sharpe_ratio = (portfolio_return - risk_free_rate) / portfolio_stddev

    # Store the results
    results[0, i] = portfolio_return
    results[1, i] = portfolio_stddev
    results[2, i] = sharpe_ratio
    results[3, i] = np.sum(weights)  # Ensure weights sum to 1

# Convert the results array into a DataFrame
results_df = pd.DataFrame(results.T, columns=['Return', 'Volatility', 'Sharpe Ratio', 'Sum of Weights'])

# Convert weights into a DataFrame for easy retrieval
weights_df = pd.DataFrame(all_weights, columns=nifty50_returns.columns)

# Extract the weights for the portfolio with the Maximum Sharpe Ratio
max_sharpe_idx = results_df['Sharpe Ratio'].idxmax()
max_sharpe_weights = weights_df.iloc[max_sharpe_idx]

# Extract the weights for the portfolio with the Minimum Volatility
min_vol_idx = results_df['Volatility'].idxmin()
min_vol_weights = weights_df.iloc[min_vol_idx]

# Display the results
print("Weights for the Portfolio with Maximum Sharpe Ratio:")
print(max_sharpe_weights)
```
